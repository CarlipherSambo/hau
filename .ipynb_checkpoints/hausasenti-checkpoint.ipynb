```python
import re
import demoji

def preprocess_tweet(text):
    # Remove URLs
    text = re.sub(r"http\S+|www\S+|https\S+", '', text, flags=re.MULTILINE)

    # Remove hashtags, mentions, and emojis
    text = re.sub(r'#\w+|\@\w+|<3|:\w+:', '', text)
    tokens = demoji.replace_with_desc(text)

    # Convert emojis to text representations
    tokens = word_tokenize(text.lower())
#     tokens_with_emojis_converted = []
#     for token in tokens:
#         if token in emoji.UNICODE_EMOJI_ENGLISH:
#             emoji_description = emoji.demojize(token)
#             emoji_tokens = emoji_description.replace(":", "").split("_")
#             tokens_with_emojis_converted.extend(emoji_tokens)
#         else:
#             tokens_with_emojis_converted.append(token)

    # Filter out custom stopwords and return the preprocessed text
    filtered_tokens = [token for token in tokens if token.isalpha() and token not in hausa_stopwords]
    return " ".join(filtered_tokens)
```

```python
train_data["cleaned_text"] = train_data["tweet"].apply(preprocess_tweet)
test_data["cleaned_text"] = test_data["tweet"].apply(preprocess_tweet)

```

```python
tfidf_vectorizer = TfidfVectorizer(max_features=5000)
X_train = tfidf_vectorizer.fit_transform(train_data["cleaned_text"]).toarray()
X_test = tfidf_vectorizer.transform(test_data["cleaned_text"]).toarray()
y_train = train_data["label"]
y_test = test_data["label"]
```

```python
from sklearn.preprocessing import LabelEncoder

# Step 5: Encode the target variable to integers
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

# Step 6: Build and train the sentiment analysis model
xgb_model = XGBClassifier()
xgb_model.fit(X_train, y_train_encoded)

```

```python
# For evaluation
y_pred_encoded = xgb_model.predict(X_test)
accuracy = accuracy_score(y_test_encoded, y_pred_encoded)
print(f"Accuracy: {accuracy:.2f}")

# For making predictions
new_text = ["wannan ai rashin mutunci ne", "jaki kawai","na gaji","toh shikenan","amma kai dan matsala ne","yaushe zanzo","da haka nacemai, amma yanzu nagane baya sona"]
preprocessed_new_text = [preprocess_tweet(text) for text in new_text]
X_new = tfidf_vectorizer.transform(preprocessed_new_text).toarray()
predictions_encoded = xgb_model.predict(X_new)

# Decode the predictions back to the original labels
predictions = label_encoder.inverse_transform(predictions_encoded)

for text, prediction in zip(new_text, predictions):
    print(f"Text: {text} | Sentiment: {prediction}")
```

```python
from hyperopt import hp, tpe, fmin
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split

# Split the data into training and validation sets
X_train_tuning, X_val_tuning, y_train_tuning, y_val_tuning = train_test_split(X_train, y_train_encoded, test_size=0.2, random_state=42)

# Define the search space for hyperparameters
param_space = {
    'learning_rate': hp.loguniform('learning_rate', -5, -2),
    'max_depth': hp.choice('max_depth', [3, 5, 7]),
    'n_estimators': hp.quniform('n_estimators', 50, 200, 1),
}
```

```python
# Define the objective function to optimize (validation accuracy)
def objective(params):
    xgb_model = XGBClassifier(
        learning_rate=params['learning_rate'],
        max_depth=params['max_depth'],
        n_estimators=int(params['n_estimators'])
    )
    xgb_model.fit(X_train_tuning, y_train_tuning)
    y_val_pred_encoded = xgb_model.predict(X_val_tuning)
    y_val_pred = label_encoder.inverse_transform(y_val_pred_encoded)
    return -accuracy_score(y_val_tuning, y_val_pred)  # Negative accuracy for minimization
```

```python
# Perform hyperparameter tuning with TPE algorithm
best = fmin(fn=objective, space=param_space, algo=tpe.suggest, max_evals=50)

# Convert the best hyperparameters to appropriate format
best_params = {
    'learning_rate': best['learning_rate'],
    'max_depth': param_space['max_depth'][best['max_depth']],
    'n_estimators': int(best['n_estimators']),
}

# Train the best model on the full training set
best_model = XGBClassifier(**best_params)
best_model.fit(X_train, y_train_encoded)
```

```python

# Evaluate the best model on the test data
y_test_pred_encoded = best_model.predict(X_test)
y_test_pred = label_encoder.inverse_transform(y_test_pred_encoded)

# Print evaluation metrics on the test data
accuracy = accuracy_score(y_test, y_test_pred)
precision = precision_score(y_test, y_test_pred, average='weighted')
recall = recall_score(y_test, y_test_pred, average='weighted')
f1 = f1_score(y_test, y_test_pred, average='weighted')

print("Evaluation Metrics for Best Model on Test Data:")
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-score: {f1:.2f}")
```

```python
import optuna
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from xgboost import XGBClassifier

# Split the data into training and validation sets
X_train_tuning, X_val_tuning, y_train_tuning, y_val_tuning = train_test_split(X_train, y_train_encoded, test_size=0.2, random_state=42)
```

```python
def objective(trial):
    # Define the search space for hyperparameters
    param = {
        'learning_rate': trial.suggest_loguniform('learning_rate', 0.01, 0.1),
        'max_depth': trial.suggest_int('max_depth', 3, 7),
        'n_estimators': trial.suggest_int('n_estimators', 50, 200),
    }

    # Initialize the XGBoost classifier
    xgb_model = XGBClassifier(**param)

    # Train the model on the reduced training set
    xgb_model.fit(X_train_tuning, y_train_tuning)

    # Predict on the validation set
    y_val_pred_encoded = xgb_model.predict(X_val_tuning)
    y_val_pred = label_encoder.inverse_transform(y_val_pred_encoded)

    # Calculate and return the validation accuracy as the objective value
    return accuracy_score(y_val_tuning, y_val_pred)
```

```python
# Perform hyperparameter tuning with Optuna
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=100, n_jobs=-1)

# Get the best hyperparameters from Optuna
best_params = study.best_params

# Train the best model on the full training set
best_model = XGBClassifier(**best_params)
best_model.fit(X_train, y_train_encoded)

# Evaluate the best model on the test data
y_test_pred_encoded = best_model.predict(X_test)
y_test_pred = label_encoder.inverse_transform(y_test_pred_encoded)

```
